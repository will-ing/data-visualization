# Data Visualization using Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Pull in data of Marvel charaters and displaying the first 5 rows

```python
marvel = pd.read_csv('./data/marvel.csv')
marvel.head()
```

### You don't need to be the smartest to be the best. 

This graph shows the best totals for superheroes from left to right and they're Speed and Intelligent attributes.

```python

sns.set()

marvel_intel = marvel[["Name", "Alignment" , "Intelligence", "Speed", "Strength", "Total"]].sort_values(by="Total", ascending=False).head(20)
name = marvel_intel["Name"]
best = sns.scatterplot(x="Name", y="Speed", label="Speed", data=marvel_intel)
best = sns.scatterplot(x="Name", y="Intelligence", label="Intelligence", data=marvel_intel)
best = sns.scatterplot(x="Name", y="Strength", label="Strength", data=marvel_intel)
best.set_xticklabels(labels=name, rotation=45)
sns.set(rc={'figure.figsize':(17,6.27)})
```
